```python
import talib
import cbpro
import pandas as pd
import requests
import numpy as np
```

```python
public_client = cbpro.PublicClient()
```

```python
hist_rates = public_client.get_product_historic_rates('BTC-USD', granularity=60)
```

```python
print(hist_rates)
len(hist_rates)
```

```python
candle_names = talib.get_function_groups()['Pattern Recognition']
candle_names
```

```python
# create dataframe with candlestick data, convert epoch time column into something useful
df = pd.DataFrame(hist_rates, columns=['time', 'low', 'high', 'open', 'close', 'volume'])
posix_time = pd.to_datetime(df['time'], unit='s')
df.insert(0, "Date", posix_time)
df.drop("time", axis = 1, inplace = True)
df
df[0:50]
```

```python
# extract Open, high, low, close
op = df['open']
hi = df['high']
lo = df['low']
cl = df['close']
```

```python
# create columns for each pattern
pd.get_option("display.max_columns")
pd.set_option("display.max_columns", None)
for candle in candle_names:
    # below is same as;
    # df["CDL3LINESTRIKE"] = talib.CDL3LINESTRIKE(op, hi, lo, cl)
    df[candle] = getattr(talib, candle)(op, hi, lo, cl)
    
df[0:50]
```

```python
# this essentially takes the original candlestick data and will find the 'best' candlestick indicator for
# each candle. If there are no indicators, it will just say "NO_PATTERN". If there are multiple candles,
# the tie will be broken based on the candleranking dictionary above. If there is one indicator for a given 
# candle, it will store that one. If the indicator is a bull indicator it will have "_Bull" appended to the
# end, if it is a bear indicator it will have "_Bear" appended to the end

df['candlestick_pattern'] = np.nan
df['candlestick_match_count'] = np.nan
row_count = 0
for index, row in df.iterrows():

    # no pattern found
    if len(row[candle_names]) - sum(row[candle_names] == 0) == 0:
        df.loc[index,'candlestick_pattern'] = "NO_PATTERN"
        df.loc[index, 'candlestick_match_count'] = 0
    # single pattern found
    elif len(row[candle_names]) - sum(row[candle_names] == 0) == 1:
        # bull pattern 100 or 200
        if any(row[candle_names].values == 100):
            candle_patterns = row[candle_names].values
            count = 0
            for candle_pattern in candle_patterns:
                if candle_pattern > 0:
                    pattern = candle_names[count] + '_Bull'
                count += 1
            df.loc[index, 'candlestick_pattern'] = pattern
            df.loc[index, 'candlestick_match_count'] = 1
        # bear pattern -100 or -200
        else:
            candle_patterns = row[candle_names].values
            count = 0
            for candle_pattern in candle_patterns:
                if candle_pattern < 0:
                    pattern = candle_names[count] + '_Bear'
                count += 1
            df.loc[index, 'candlestick_pattern'] = pattern
            df.loc[index, 'candlestick_match_count'] = 1
    # multiple patterns matched -- select best performance
    else:
        patterns = []
        candle_patterns = row[candle_names].values
        count = 0
        for candle_pattern in candle_patterns:
            if candle_pattern == 100:
                pattern = candle_names[count]
                patterns.append(pattern)
            elif candle_pattern == -100:
                pattern = candle_names[count]
                patterns.append(pattern)
            count += 1
        # filter out pattern names from bool list of values
        containers = []
        for pattern in patterns:
            if row[pattern] > 0:
                containers.append(pattern + '_Bull')
            else:
                containers.append(pattern + '_Bear')
        
        # this is really jank but hey it works
        count = 0
        print(row_count)
        ranks = []
        for container in containers:
            ranks.append(candle_rankings[container])
        ranks.sort()
        winning_container_rank = ranks[0]
        
        for container in containers:
            print(candle_rankings[container])
            if candle_rankings[container] == winning_container_rank:
                winning_container = container
         
        print(winning_container_rank)
        print()
            
        
        df.loc[index, 'candlestick_pattern'] = winning_container
        df.loc[index, 'candlestick_match_count'] = 1
        row_count += 1
        
       
# clean up candle columns
df.drop(candle_names, axis = 1, inplace = True)
```
